```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.table import Table
from astropy.io import fits
from scipy.optimize import curve_fit
from clevar.match import output_matched_catalog
from clevar.catalog import ClCatalog
```

```python
inpath = "/sps/lsst/groups/clusters/amico_validation_project/catalogs/"
rm_inpath = '/sps/lsst/users/tguillem/DESC/desc_april_2022/cluster_challenge/clevar_catalogs/redmapper/full_pmem/cosmoDC2_v1.1.4_redmapper_v0.8.1/'
```

```python
am_path = inpath+'AMICO/amico_cats/amico_map_associations_flxzb_mag/mag_i/Catalog.fits'
rm_path = rm_inpath + 'Catalog.fits'
```

```python
am = Table.read(am_path)
rm = Table.read(rm_path)
```

```python
bins = np.linspace(0,100,50)
```

```python
plt.hist(am['mass'], bins = bins, label = 'AMICO', histtype = 'step', density = True)
plt.hist(rm['mass'], bins = bins, label = 'redMaPPer', histtype = 'step', density = True)
plt.xlim([0,80])
plt.xlabel('$\lambda^*$(AMICO) & $\lambda$(redMaPPer)', fontsize = 15)
plt.ylabel('Number of clusters', fontsize = 15)
plt.legend()
plt.savefig('/pbs/home/n/namourou/test_jupyter/cluster_challenge/plots/richness_diff.png', format='png', transparent=True)
plt.show()
```
