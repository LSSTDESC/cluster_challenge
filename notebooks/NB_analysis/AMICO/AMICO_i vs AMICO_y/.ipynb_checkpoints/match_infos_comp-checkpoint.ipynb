```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.table import Table
from astropy.io import fits
from scipy.optimize import curve_fit
from clevar.match import output_matched_catalog
from clevar.catalog import ClCatalog
```

```python
inpath = "/sps/lsst/groups/clusters/amico_validation_project/catalogs/matching_cats/amico_cosmoDC2/mag_i/0.0/"
```

```python
c1 = Table.read(inpath + 'c1_mb.fits')
c2 = Table.read(inpath + 'c2_mb.fits')
c_merged_12 = Table.read(inpath + 'output_catalog_mb.fits')
```

```python
plt.hist(c1['z'], bins = 50, label = 'AMICO$_i$', histtype = 'step')
plt.hist(c2['z'], bins = 50, label = 'cosmoDC2', histtype = 'step')
#plt.xlim([0,50])
plt.xlabel('z', fontsize = 13)
plt.ylabel('Number of clusters', fontsize = 13)
plt.legend()
plt.show()
```

```python
plt.hist(c_merged_12['cat1_z'], bins = 50, label = 'AMICO$_i$', histtype = 'step')
plt.hist(c_merged_12['cat2_z'], bins = 50, label = 'cosmoDC2', histtype = 'step')
#plt.xlim([0,50])
plt.xlabel('z', fontsize = 13)
plt.ylabel('Number of clusters', fontsize = 13)
plt.legend()
plt.show()
```

```python
plt.scatter(c_merged_12['cat1_z'],c_merged_12['cat2_z'], s=2, label = 'fs=0')
plt.xlabel('$z_{AMICO}$', fontsize = 13)
plt.ylabel('$z_{cosmoDC2}$', fontsize = 13)
plt.legend()
```

### Multiple scatter plots

```python
inpatha = "/sps/lsst/groups/clusters/amico_validation_project/catalogs/matching_cats/amico_cosmoDC2/mag_i/0.05/"
inpathb = "/sps/lsst/groups/clusters/amico_validation_project/catalogs/matching_cats/amico_cosmoDC2/mag_i/0.1/"
inpathc = "/sps/lsst/groups/clusters/amico_validation_project/catalogs/matching_cats/amico_cosmoDC2/mag_i/0.2/"
```

```python
c1a = Table.read(inpatha + 'c1_mb.fits')
c2a = Table.read(inpatha + 'c2_mb.fits')
c_merged_12a = Table.read(inpatha + 'output_catalog_mb.fits')

c1b = Table.read(inpathb + 'c1_mb.fits')
c2b = Table.read(inpathb + 'c2_mb.fits')
c_merged_12b = Table.read(inpathb + 'output_catalog_mb.fits')

c1c = Table.read(inpathc + 'c1_mb.fits')
c2c = Table.read(inpathc + 'c2_mb.fits')
c_merged_12c = Table.read(inpathc + 'output_catalog_mb.fits')
```

```python
plt.scatter(c_merged_12['cat1_z'],c_merged_12['cat2_z'], s=2, label = 'fs=0')
plt.scatter(c_merged_12a['cat1_z'],c_merged_12a['cat2_z'], s=2, label = 'fs=0.05')
plt.scatter(c_merged_12b['cat1_z'],c_merged_12b['cat2_z'], s=2, label = 'fs=0.1')
plt.scatter(c_merged_12c['cat1_z'],c_merged_12c['cat2_z'], s=2, label = 'fs=0.2')
plt.xlabel('$z_{AMICO}$', fontsize = 13)
plt.ylabel('$z_{cosmoDC2}$', fontsize = 13)
plt.legend()
```

```python
plt.hist(c_merged_12['cat1_z'], bins = 50, label = 'fs=0', histtype = 'step')
plt.hist(c_merged_12a['cat1_z'], bins = 50, label = 'fs=0.05', histtype = 'step')
plt.hist(c_merged_12b['cat1_z'], bins = 50, label = 'fs=0.1', histtype = 'step')
plt.hist(c_merged_12c['cat1_z'], bins = 50, label = 'fs=0.2', histtype = 'step')

#plt.xlim([0,50])
plt.xlabel('z', fontsize = 13)
plt.ylabel('Number of clusters', fontsize = 13)
plt.legend()
plt.show()
```

```python
plt.hist(c_merged_12['cat1_mass'], bins = 50, label = 'fs=0', histtype = 'step')
plt.hist(c_merged_12a['cat1_mass'], bins = 50, label = 'fs=0.05', histtype = 'step')
plt.hist(c_merged_12b['cat1_mass'], bins = 50, label = 'fs=0.1', histtype = 'step')
plt.hist(c_merged_12c['cat1_mass'], bins = 50, label = 'fs=0.2', histtype = 'step')
plt.xlim([0,100])
#plt.xlim([0,50])
plt.xlabel('z', fontsize = 13)
plt.ylabel('Number of clusters', fontsize = 13)
plt.legend()
plt.show()
```
