```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.table import Table
from astropy.io import fits
```

### Loading Matching's results

```python
cl_inpath = '/sps/lsst/groups/clusters/amico_validation_project/catalogs/AMICO/raw_amico_cats/cosmoDC2_small_photoz_flexzboost_v1_iband/map_detections_refined_noBuffer_all_noDoubles.fits'
```

```python
amico_cl_data = Table.read(cl_inpath)['ID', 'Xphys', 'Yphys', 'Zphys','LAMBSTAR', 'SN_NO_CLUSTER', 'SN', 'UID', 'TILE']
```

```python
amico_cl_data
```

```python
min(amico_cl_data['SN'])
```

```python
plt.hist(amico_cl_data['SN'], bins = 25)
```

```python
z_l = [.1,.3,.5,.7,.9,1.1]
```

```python
colors = ['b', 'g', 'm','r','tab:orange']
```

```python
for i in range(1,len(z_l)):
    z_min = z_l[i-1]
    z_max = z_l[i]
    amico_cl_cdt = amico_cl_data[(amico_cl_data['Zphys']>z_min)*(amico_cl_data['Zphys']>z_max)]
    plt.scatter(amico_cl_cdt['LAMBSTAR'], amico_cl_cdt['SN'], label = str(z_min) + '<z<' + str(z_max), s=5, color = colors[i-1], alpha = .3)
plt.ylabel('S/N', fontsize = 15)
plt.xlabel('$\lambda_{*}$', fontsize = 15)
plt.ylim([2,12])
plt.xlim([0,150])
plt.legend()
```

```python
z_l2 = [1.1, 1.3, 1.5, 1.9]
for i in range(1,len(z_l2)):
    z_min = z_l2[i-1]
    z_max = z_l2[i]
    amico_cl_cdt = amico_cl_data[(amico_cl_data['Zphys']>z_min)*(amico_cl_data['Zphys']>z_max)]
    plt.scatter(amico_cl_cdt['LAMBSTAR'], amico_cl_cdt['SN'], label = str(z_min) + '<z<' + str(z_max), s=5, alpha = .3)
plt.ylabel('S/N', fontsize = 15)
plt.xlabel('$\lambda_{*}$', fontsize = 15)
plt.ylim([2,12])
plt.xlim([0,150])
plt.legend()
```

```python
for i in range(1,len(z_l)):
    z_min = z_l[i-1]
    z_max = z_l[i]
    amico_cl_cdt = amico_cl_data[(amico_cl_data['SN']>3.5)*(amico_cl_data['Zphys']>z_min)*(amico_cl_data['Zphys']>z_max)]
    plt.scatter(amico_cl_cdt['AMP'], amico_cl_cdt['LAMBSTAR'], label = str(z_min) + '<z<' + str(z_max), s=5, color = colors[i-1])
plt.xlabel('A', fontsize = 15)
plt.ylabel('$\lambda_{*}$', fontsize = 15)
plt.ylim([0,120])
plt.xlim([0,12])
plt.legend()
```
