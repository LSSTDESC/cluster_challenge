### Load modules

```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.table import Table
```

## Clusters' distributions

```python
cl_inpath = '/sps/lsst/groups/clusters/amico_validation_project/catalogs/AMICO/raw_amico_cats/cosmoDC2_photoz_flexzboost_v1_iband/map_detections_refined_noBuffer_all_noDoubles.fits'
amico_cl_data = Table.read(cl_inpath)['ID', 'Xphys', 'Yphys', 'Zphys','LAMBSTAR', 'SN_NO_CLUSTER', 'SN', 'UID', 'TILE', 'LAMB', 'AMP']  
```

```python
amico_cl_data
```

### AMICO footprint

```python
plt.scatter(amico_cl_data['Xphys'],amico_cl_data['Yphys'],s=0.1, alpha = .5)
plt.xlabel('ra', fontsize = 15)
plt.ylabel('dec', fontsize = 15)
plt.savefig('/pbs/home/n/namourou/test_jupyter/cluster_challenge/plots/all_map.png', format='png', transparent=True)
plt.show()
```

### Amplitude distribution

```python
amico_cl_cdt1 = amico_cl_data[(amico_cl_data['SN']>2)]
amico_cl_cdt2 = amico_cl_data[(amico_cl_data['SN']>4)]
plt.hist(amico_cl_cdt1['AMP'], bins = 100, alpha =.75, label = 'SN>2')
plt.hist(amico_cl_cdt2['AMP'], bins =  100, alpha = .75, label = 'SN>4')
plt.xlabel('AMP', fontsize = 15)
plt.ylabel('count', fontsize = 15)
#plt.ylim([0,120])
#plt.xlim([0,12])
plt.legend()
```

### Richness distribution

```python
plt.hist(amico_cl_cdt1['LAMB'], bins = 200, alpha =.75, label = 'SN>2')
plt.hist(amico_cl_cdt2['LAMB'], bins =  200, alpha = .75, label = 'SN>4')
plt.xlabel('$\lambda$', fontsize = 15)
plt.ylabel('count', fontsize = 15)
#plt.ylim([0,120])
#plt.xlim([0,12])
plt.xscale('log')
plt.legend()
```

### Intrinsic richness distribution

```python
plt.hist(amico_cl_cdt1['LAMBSTAR'], bins = 200, alpha =.75, label = 'SN>2')
plt.hist(amico_cl_cdt2['LAMBSTAR'], bins =  200, alpha = .75, label = 'SN>4')
plt.xlabel('$\lambda_*$', fontsize = 15)
plt.ylabel('count', fontsize = 15)
#plt.ylim([0,120])
#plt.xlim([0,100])
plt.xscale('log')
plt.legend()
```

### Redshift distribution

```python
plt.hist(amico_cl_data['Zphys'], bins = 100)
#plt.hist(amico_cl_cdt2['Zphys'], bins =  50, label = 'SN>4')
plt.xlabel('$z$', fontsize = 15)
plt.ylabel("Number of clusters", fontsize = 15)
#plt.ylim([0,120])
plt.xlim([0,2])
#plt.legend()
plt.savefig('/pbs/home/n/namourou/test_jupyter/cluster_challenge/plots/am_zdistrib.png', format='png', transparent=True)
plt.show()
```

### SNR distribution

```python
plt.hist(amico_cl_data['SN'], bins = 50, label = 'SN>2')
plt.xlabel('SN', fontsize = 15)
plt.ylabel('count', fontsize = 15)
#plt.ylim([0,120])
#plt.xlim([0,12])
```

### Intrinsic richness vs z vs SN/R

```python
plt.scatter(amico_cl_data['Zphys'],amico_cl_data['LAMBSTAR'],s=2, c=amico_cl_data['SN'], cmap = 'jet',vmax = 6)
cbar = plt.colorbar()
cbar.set_label(label = 'SNR', size = 13)
plt.xlabel('$z$', fontsize = 13)
plt.ylabel('$\lambda^*$', fontsize = 13)
plt.ylim([0,100])
plt.xlim([0,2])
plt.savefig('/pbs/home/n/namourou/test_jupyter/cluster_challenge/plots/lambdazsn.png', format='png', transparent=True)
```
