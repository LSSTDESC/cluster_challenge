# p_mem distributions for AMICO

```python
import matplotlib.pyplot as plt
import numpy as np
from astropy.table import Table, hstack, vstack
from clevar.catalog import ClCatalog
from numpy.ma import masked
import sys
from clevar.match import output_matched_catalog
```

```python
am_inpath = '/sps/lsst/groups/clusters/amico_validation_project/catalogs/AMICO/amico_cats/amico_map_associations_flxzb_mag/mag_i/Catalog_members.fits' 
#cdc_inpath = "/sps/lsst/users/tguillem/DESC/desc_april_2022/cluster_challenge/clevar_catalogs/cosmoDC2/m200c_gt_13.0/cosmoDC2_v1.1.4_image/Catalog_members.fits"
am_mb = Table.read(am_inpath)
#cdc_mb = Table.read(cdc_inpath)
```

## Not matched clusters

```python
#plt.hist(am_mb['pmem'], bins = np.logspace(-3,0,1000))
plt.hist(am_mb['pmem'], bins = np.logspace(-3,0,100))
plt.xscale('log')
plt.xlim([10**-3,2])
plt.xlabel('pmem', fontsize = 13)
plt.ylabel('Number of galaxies', fontsize = 13)
plt.title('Distribution of AMICO pmem (before matching)', fontsize = 13)
plt.show()
```

```python
print('µ = ', np.mean(am_mb['pmem']), '\nσ = ', np.std(am_mb['pmem']))
```

## Matched clusters

```python
inpath = "/sps/lsst/groups/clusters/amico_validation_project/catalogs/matching_cats/amico_cosmoDC2/mag_i/"
outpath = inpath
print('reading tables')
c1_mb = Table.read(inpath + 'c1_p_members.fits')
c2_mb = Table.read(inpath + 'c2_p_members.fits')
```

```python
plt.hist(c1_mb['pmem'], bins = np.logspace(-3,0,100))
plt.xscale('log')
plt.xlim([10**-2.5,2])
plt.xlabel('pmem', fontsize = 13)
plt.ylabel('Number of members', fontsize = 13)
plt.title('Distribution of members pmem for matched clusters')
```
