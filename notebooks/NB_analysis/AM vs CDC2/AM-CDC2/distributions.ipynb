# Distributions

```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.table import Table
from astropy.io import fits
```

```python
inpath = "/sps/lsst/groups/clusters/amico_validation_project/catalogs/matching_cats/amico_cosmoDC2/mag_i/"
```

```python
c1 = Table.read(inpath + 'c1_p.fits')
c2 = Table.read(inpath + 'c2_p.fits')
c_merged_12 = Table.read(inpath + 'output_catalog_p.fits')
```

## Redshift Distributions

### Full AMICO & cosmoDC2 catalogs

```python
plt.hist(c1['z'], bins = 50, label = 'AMICO$_i$', histtype = 'step')
plt.hist(c2['z'], bins = 50, label = 'cosmoDC2', histtype = 'step')
#plt.xlim([0,50])
plt.xlabel('z', fontsize = 13)
plt.ylabel('Number of clusters', fontsize = 13)
plt.title('Not matched clusters')
plt.legend()
plt.show()
```

### Matched AMICO & cosmoDC2

```python
plt.hist(c_merged_12['cat1_z'], bins = 50, label = 'AMICO$_i$', histtype = 'step')
plt.hist(c_merged_12['cat2_z'], bins = 50, label = 'cosmoDC2', histtype = 'step')
#plt.xlim([0,50])
plt.xlabel('z', fontsize = 13)
plt.ylabel('Number of clusters', fontsize = 13)
plt.title('Matched clusters')
plt.legend()
plt.show()
```

## Correlation z_AMICO & z_cosmoDC2

```python
plt.scatter(c_merged_12['cat1_z'],c_merged_12['cat2_z'], s=2, label = 'fs=0')
plt.xlabel('$z_{AMICO}$', fontsize = 13)
plt.ylabel('$z_{cosmoDC2}$', fontsize = 13)
plt.title('Matched clusters')
plt.legend()
```
