# Intrinsic richness distributions

```python
import matplotlib.pyplot as plt
import numpy as np
from astropy.table import Table
```

## Loading catalogs

```python
all_amico_data_members = Table.read('/sps/lsst/groups/clusters/amico_validation_project/catalogs/amico_cats/new_amico_map_associations_cut/all_maps_cut.fits')
```

```python
all_amico_data_members
```

```python
cl_inpath = '/sps/lsst/groups/clusters/amico_validation_project/catalogs/130323/cosmoDC2_small_photoz_flexzboost_v0/map_detections_refined_noBuffer_all_noDoubles.fits'
amico_cl_data = Table.read(cl_inpath)['ID', 'Xphys', 'Yphys', 'Zphys','LAMBSTAR', 'SN_NO_CLUSTER', 'SN', 'UID', 'TILE', 'LAMB', 'AMP']  
```

```python
amico_cl_data
```

## Full catalog $\lambda^*$ distribution

```python
plt.hist(amico_cl_data['LAMBSTAR'], bins = 20)
plt.ylabel('count', fontsize = 15)
plt.xlabel('$\lambda_*$', fontsize = 15)
```

## $z$ distribution in $\lambda^*$ bins

```python
amico_cl_cdt1 = amico_cl_data[(amico_cl_data['LAMBSTAR']>0)*(amico_cl_data['LAMBSTAR']<10)]
amico_cl_cdt2 = amico_cl_data[(amico_cl_data['LAMBSTAR']>10)*(amico_cl_data['LAMBSTAR']<20)]
amico_cl_cdt3 = amico_cl_data[(amico_cl_data['LAMBSTAR']>20)*(amico_cl_data['LAMBSTAR']<30)]
amico_cl_cdt4 = amico_cl_data[(amico_cl_data['LAMBSTAR']>30)*(amico_cl_data['LAMBSTAR']<40)]
amico_cl_cdt5 = amico_cl_data[(amico_cl_data['LAMBSTAR']>40)*(amico_cl_data['LAMBSTAR']<50)]

plt.hist(amico_cl_cdt1['Zphys'], bins = 20, alpha = .75, label = '0<$\lambda_*$<10', density = True)
plt.hist(amico_cl_cdt2['Zphys'], bins = 20, alpha = .75, label = '10<$\lambda_*$<20', density = True)
plt.hist(amico_cl_cdt3['Zphys'], bins = 20, alpha = .75, label = '20<$\lambda_*$<30', density = True)
#plt.hist(amico_cl_cdt4['Zphys'], bins = 20, alpha = .75, label = '30<$\lambda_*$<40', density = True)
#plt.hist(amico_cl_cdt5['Zphys'], bins = 20, alpha = .75, label = '40<$\lambda_*$<50', density = True)
plt.xlabel('redshift', fontsize = 15)
plt.ylabel('count', fontsize = 15)
#plt.ylim([0,120])
#plt.xlim([0,12])
plt.legend()
```

```python
bins = np.linspace(0,50, 50)
```

```python
amico_cl_cdt1 = amico_cl_data[(amico_cl_data['Zphys']>0)*(amico_cl_data['Zphys']<.5)]
amico_cl_cdt2 = amico_cl_data[(amico_cl_data['Zphys']>.5)*(amico_cl_data['Zphys']<.8)]
amico_cl_cdt3 = amico_cl_data[(amico_cl_data['Zphys']>.8)*(amico_cl_data['Zphys']<1.0)]
amico_cl_cdt4 = amico_cl_data[(amico_cl_data['Zphys']>1.0)*(amico_cl_data['Zphys']<1.2)]
amico_cl_cdt5 = amico_cl_data[(amico_cl_data['Zphys']>1.2)*(amico_cl_data['Zphys']<1.5)]

plt.hist(amico_cl_cdt1['LAMBSTAR'], bins = bins, alpha = .5, label = '0<z<0.5', density = True, histtype = 'step')
plt.hist(amico_cl_cdt2['LAMBSTAR'], bins = bins, alpha = .5, label = '0.5<z<0.8', density = True, histtype = 'step')
plt.hist(amico_cl_cdt3['LAMBSTAR'], bins = bins, alpha = .5, label = '0.8<z<1.0', density = True, histtype = 'step')
plt.hist(amico_cl_cdt4['LAMBSTAR'],  color = 'cyan', bins = bins, alpha = .5, label = '1.0<z<1.2', density = True, histtype = 'step')
plt.hist(amico_cl_cdt5['LAMBSTAR'], bins = bins, alpha = .5, label = '1.2<z<1.5', density = True, histtype = 'step')
plt.xlabel('$\lambda_*$', fontsize = 15)
plt.ylabel('count (normalized)', fontsize = 15)
#plt.ylim([0,120])
plt.xlim([0,50])
plt.legend()
```
