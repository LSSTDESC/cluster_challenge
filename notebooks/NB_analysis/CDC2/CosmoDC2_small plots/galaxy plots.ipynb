```python
import pandas as pd
galaxy_data_df = pd.DataFrame(galaxy_data)
bin_corners = np.linspace(min(galaxy_data_df['redshift']),max(galaxy_data_df['redshift']),10)
bin_list = [[bin_corners[i],bin_corners[i+1]] for i in range(len(bin_corners)-1)]
mag_r = []
mag_i = []
mag_z = []
for bin in bin_list:
    mask = (galaxy_data_df['redshift'] < bin[1]) & (galaxy_data_df['redshift'] > bin[0])
    mag_r_cut = galaxy_data_df['mag_r'][mask]
    mag_i_cut = galaxy_data_df['mag_i'][mask]
    mag_z_cut = galaxy_data_df['mag_z'][mask]
    mag_r.append(np.mean(mag_r_cut)) #Append new point to list
    mag_i.append(np.mean(mag_i_cut))
    mag_z.append(np.mean(mag_z_cut))
mid_bin = np.mean(bin_list,  axis=1)
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize = (20,6))
ax1.errorbar(mid_bin, mag_r, marker = '+', linestyle = 'none',markersize = 10)
ax2.errorbar(mid_bin, mag_i, marker = '+', linestyle = 'none',markersize = 10)
ax3.errorbar(mid_bin, mag_z, marker = '+', linestyle = 'none',markersize = 10)
ax_l = [ax1,ax2,ax3]
for ax in ax_l:
    ax.set_ylim(16.5,19)
    ax.set_xlabel('redshift', fontsize = 20)
ax1.set_ylabel('$mag_r$', fontsize = 20)
ax2.set_ylabel('$mag_i$', fontsize = 20)
ax3.set_ylabel('$mag_z$', fontsize = 20)
plt.savefig('subplot2')
```
