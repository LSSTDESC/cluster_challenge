```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.table import Table
from astropy.io import fits
from scipy.optimize import curve_fit
from clevar.match import output_matched_catalog
from clevar.catalog import ClCatalog
```

```python
inpath = "/sps/lsst/groups/clusters/amico_validation_project/catalogs/"
rm_inpath = '/sps/lsst/users/tguillem/DESC/desc_april_2022/cluster_challenge/clevar_catalogs/redmapper/full_pmem/cosmoDC2_v1.1.4_redmapper_v0.8.1/'
```

```python
am_path = inpath+'AMICO/amico_cats/amico_map_associations_flxzb_mag/mag_y/Catalog.fits'
rm_path = rm_inpath + 'Catalog.fits'
```

```python
am = Table.read(am_path)
rm = Table.read(rm_path)
```

```python
bins = np.linspace(0,50,50)
```

```python
plt.hist(am['mass'], bins = bins, label = 'amico', histtype = 'step')
plt.hist(rm['mass'], bins = bins, label = 'Redmapper', histtype = 'step')
plt.xlim([0,50])
plt.xlabel('$\lambda$', fontsize = 15)
plt.ylabel('Number of clusters', fontsize = 15)
plt.legend()
plt.show()
```

```python
plt.hist(am['z'], bins = 25, label = 'amico', histtype = 'step')
plt.hist(rm['z'], bins = 25, label = 'Redmapper', histtype = 'step')
plt.hist(rm['z_true'], bins = 25, label = 'Redmapper_z_true', histtype = 'step')
plt.xlim([0,2])
plt.xlabel('z', fontsize = 15)
plt.ylabel('Number of clusters', fontsize = 15)
plt.legend()
plt.show()
```
