### Reproducing Euclid plots

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Overmerging

```python
bo_x = [13.7,13.95,14.2,14.45,14.7,14.95,15.20,15.35]
bo_y = [0.15,0.112,0.13,0.058,1,0,0,0]
blo_x = [bo_x[i] - 0.02 for i in range(len(bo_x))]
blo_y = [0.195,0.17,0.19,0.2,0.21,0,0,0]
go_x = [blo_x[i] - 0.02 for i in range(len(bo_x))]
go_y = [0.205,0.205,0.23,0.29,0.285,0.25,0.27,0]
ro_x = [go_x[i] - 0.02 for i in range(len(bo_x))]
ro_y = [0.21, 0.23,0.25,0.26,0.4,0.35,0.35,1]
```

```python
plt.plot(bo_x, bo_y, marker = 'o', color = 'black', linestyle = 'None', label = '1.5 < z < 2.0')
plt.plot(blo_x, blo_y, marker = 'o', color = 'blue', linestyle = 'None', label = '1.0 < z < 1.5')
plt.plot(go_x, go_y, marker = 'o', color = 'green', linestyle = 'None', label = '0.5 < z < 1.0')
plt.plot(ro_x, ro_y, marker = 'o', color = 'red', linestyle = 'None', label = '0.0 < z < 0.5')
plt.xlabel('log(M/$M_{\odot}$)', fontsize = 13)
plt.ylabel('Overmerging rate', fontsize = 13)
plt.ylim([10**-2,2])
plt.yscale('log')
plt.legend()
```

```python
bf_x = [13.7,13.95,14.2,14.45,14.7,14.95,15.2]
bf_y = [0.016,0.05,0.09,0.12,1,1,1]
blf_x = [bf_x[i] - 0.02 for i in range(len(bf_x))]
blf_y = [0.06,0.09,0.19,0.35,0.6,0.8,1]
gf_x = [blf_x[i] - 0.02 for i in range(len(bf_x))]
gf_y = [0.12,0.20,0.31,0.6,0.8,0.95,1]
rf_x = [gf_x[i] - 0.02 for i in range(len(bf_x))]
rf_y = [0.16,0.3,0.5,0.7,0.9,1,1]
```

```python
plt.plot(bf_x, bf_y, marker = 'o', color = 'black', linestyle = 'None', label = '1.5 < z < 2.0')
plt.plot(blf_x, blf_y, marker = 'o', color = 'blue', linestyle = 'None', label = '1.0 < z < 1.5')
plt.plot(gf_x, gf_y, marker = 'o', color = 'green', linestyle = 'None', label = '0.5 < z < 1.0')
plt.plot(rf_x, rf_y, marker = 'o', color = 'red', linestyle = 'None', label = '0.0 < z < 0.5')
plt.yscale('log')
plt.xlabel('log(M/$M_{\odot}$)', fontsize = 13)
plt.ylabel('Fragmentation rate', fontsize = 13)
plt.ylim([10**-2,10])
plt.legend()
```
